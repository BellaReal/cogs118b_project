```python
import pandas as pd
import numpy as np
import missingno as msno 
import seaborn as sns
```

```python
df = pd.read_csv("owid-covid-data.csv", low_memory=False)
df.head()
```

```python
msno.matrix(df);
```

```python
msno.bar(df);
```

```python
msno.heatmap(df);
```

```python
sns.countplot(df, y='location');
sns.set(rc={'figure.figsize':(15,30)})
```

```python
df
```

```python
sns.lineplot(data=df, x="date", y="new_cases");
sns.set(rc={'figure.figsize':(10,8)});
```

## United States Subsection

```python
df_us = df[df['location']=='United States']
df_us
```

```python
sns.lineplot(data=df_us, x="date", y="new_deaths_smoothed");
sns.set(rc={'figure.figsize':(8,5)});
```

```python
big_cases = df_us[df_us['new_death_smoothed'] > 700000]
big_cases
```

```python
sns.lineplot(data=df_us, x="date", y="total_cases");
```
